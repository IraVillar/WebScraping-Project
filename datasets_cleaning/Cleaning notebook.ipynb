```python
import pandas as pd
import re

def reader(csv):
    csv = pd.read_csv(csv,header=None)
    
    return csv

def cleandataframe(df):
    df.columns=['Item', 'Price', 'Username']
    df.drop_duplicates(subset=None, keep='first', inplace=True)
    df.reset_index(drop=True, inplace = True)
    df['Price'] = df['Price'].apply(lambda x: x.replace(',',"").strip())
    df['Price'] = df['Price'].apply(lambda x: float(x.replace('PHP',"")))
    
    return df
```

```python
realestate = reader('realestate.csv')
realestate = cleandataframe(realestate)
```

```python
cars = reader('cars.csv')
cars = cleandataframe(cars)
```

```python
mobile = reader('mobile.csv')
mobile = cleandataframe(mobile)
```
