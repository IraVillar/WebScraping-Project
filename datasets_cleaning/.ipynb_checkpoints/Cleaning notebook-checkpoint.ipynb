```python
#Functions and imports
import pandas as pd
import seaborn as sns
import numpy as np
import plotly
from matplotlib import pyplot as plt
from plotly.graph_objs import Scatter
from plotly.graph_objs import Figure, Histogram, Layout

def reader(csv):
    csv = pd.read_csv(csv,header=None)
    
    return csv

def cleandataframe3(df):
    df.columns=['Item', 'Price', 'Username']
    df.drop_duplicates(subset=None, keep='first', inplace=True)
    df.dropna()
    df['Price'] = df['Price'].apply(lambda x: x.replace(',',"").strip())
    df['Price'] = df['Price'].apply(lambda x: float(x.replace('PHP',"")))
    df = df.sort_values('Price',ascending=True)
    df.reset_index(drop=True, inplace = True)
    df['Price in USD'] = df['Price'] / 50
    df['index'] = df.index
    
    return df

def cleandataframe(df):
    df.columns=['Item', 'Price', 'Username', 'Status']
    df.drop_duplicates(subset=None, keep='first', inplace=True)
    df.dropna()
    df['Price'] = df['Price'].apply(lambda x: x.replace(',',"").strip())
    df['Price'] = df['Price'].apply(lambda x: float(x.replace('PHP',"")))
#     df = df.sort_values('Price',ascending=True)
    df.reset_index(drop=True, inplace = True)
    df['Price in USD'] = df['Price'] / 50
    df['index'] = df.index
    
    return df
```

```python
#Reading
assistive = reader('assistive.csv')
realestate = reader('3realestate.csv')
cars = reader('cars.csv')
mobile = reader('mobile.csv')
antiques = reader('antiques.csv')
babies = reader('babies.csv')
books = reader('books.csv')
craft = reader('craft.csv')
electronics = reader('electronics.csv')
car_parts = reader('car_parts.csv')
health_and_beauty = reader('hnb.csv')
hnf = reader('hnf.csv')
hs = reader('home_services.csv')
luxury = reader('luxury.csv')
mens = reader('mens.csv')
motorbikes = reader('motorbikes.csv')
music = reader('music.csv')
sports = reader('sports.csv')
textbooks = reader('textbooks.csv')
toys = reader('toys.csv')
travel_essentials = reader('travel_essentials.csv')
vehicle_rental = reader('vehicle_rental.csv')
videogames = reader('videogames.csv')
womens = reader('womens.csv')
photography = reader('photography.csv')
lifestyle = reader('lifestyle.csv')
```

```python
#Cleaning
assistive = cleandataframe(assistive)
lifestyle= cleandataframe(lifestyle)
photography = cleandataframe(photography)
realestate = cleandataframe3(realestate)
antiques = cleandataframe(antiques)
babies = cleandataframe(babies)
books = cleandataframe(books)
car_parts = cleandataframe(car_parts)
craft = cleandataframe(craft)
electronics = cleandataframe(electronics)
health_and_beauty = cleandataframe(health_and_beauty)
home_and_furniture = cleandataframe(hnf)
home_services = cleandataframe(hs)
luxury = cleandataframe(luxury)
mens = cleandataframe(mens)
motorbikes = cleandataframe(motorbikes)
music = cleandataframe(music)
sports = cleandataframe(sports)
textbooks = cleandataframe(textbooks)
toys = cleandataframe(toys)
cars = cleandataframe(cars)
mobile = cleandataframe(mobile)
travel_essentials = cleandataframe(travel_essentials)
vehicle_rental = cleandataframe(vehicle_rental)
videogames = cleandataframe(videogames)
womens = cleandataframe(womens)
```

```python
#merging
clothes = pd.concat([mens, womens]).reset_index(drop=True)
```
